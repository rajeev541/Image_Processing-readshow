```python
img1=cv2.imread('/content/drive/MyDrive/img3.jfif')
```

```python
img2=cv2.imread('/content/drive/MyDrive/img2.jpg')
```

```python
img1=cv2.resize(img1,(183,183))
```

```python
img2=cv2.resize(img2,(183,183))
```

```python
print(img1.shape)
```

```python
print(img2.shape)
```

```python
dst_add=cv2.addWeighted(img1,0.3,img2,0.9,0)
```

```python
plt.imshow(dst_add)
```

```python
sub=cv2.subtract(img1,img2)
```

```python
plt.imshow(sub)
```

```python
mul=cv2.multiply(img1,img2)
```

```python
plt.imshow(mul)
```

```python
div=cv2.divide(img1,img2)
```

```python
plt.imshow(div)
```

```python
dst_and = cv2.bitwise_and(img1, img2, mask = None)
```

```python
plt.imshow(dst_and)
```

```python
dst_and = cv2.bitwise_or(img1, img2, mask = None)
```

```python
plt.imshow(dst_and)
```

```python
dst_xor = cv2.bitwise_xor(img1, img2, mask = None)
```

```python
plt.imshow(dst_xor)
```

```python
dst_not = cv2.bitwise_not(img1, mask = None)
```

```python
plt.imshow(dst_not)
```

```python
dst_not = cv2.bitwise_not(img2, mask = None)
```

```python
plt.imshow(dst_not)
```
